```python
x = np.arange(0, 10)
```

```python
fig, ax = plt.subplots(figsize=(10, 4))

ax.plot(x, x**2, x, x**3, lw=1)

ax.set_xticks([1, 2, 3, 4, 5])
ax.set_xticklabels([r'$\alpha$', r'$\beta$', r'$\gamma$', r'$\delta$', r'$\epsilon$'], fontsize=18)

yticks = [0, 50, 100, 150]
ax.set_yticks(yticks)
ax.set_yticklabels(["$%.1f$" % y for y in yticks], fontsize=18); # use LaTeX formatted labels
```

```python
fig, ax = plt.subplots(1, 1)
      
ax.plot(x, x**2, x, np.exp(x))
ax.set_title("scientific notation")

ax.set_yticks([0, 50, 100, 150])

formatter = ticker.ScalarFormatter(useMathText=True)
formatter.set_scientific(True)
formatter.set_powerlimits((-1,1)) 
ax.yaxis.set_major_formatter(formatter)
```

```python
fig, ax1 = plt.subplots()

ax1.plot(x, x**2, lw=2, color="blue")
ax1.set_ylabel(r"area $(m^2)$", fontsize=18, color="blue")
for label in ax1.get_yticklabels():
    label.set_color("blue")
    
ax2 = ax1.twinx()
ax2.plot(x, x**3, lw=2, color="red")
ax2.set_ylabel(r"volume $(m^3)$", fontsize=18, color="red")
for label in ax2.get_yticklabels():
    label.set_color("red")
```

```python
x = np.arange(0, 10)

fig, (axes1, axes2) = plt.subplots(1, 2, figsize=(10,4))
      
line1, = axes1.plot(x, x**2,  label ="Line 1")
line2, = axes1.plot(x, np.exp(x),  label ="Line 2")

# 
first_legend = axes1.legend(handles =[line1], loc = (0.1,0.88))
two_legend = axes1.legend(handles =[line2], loc = (0.1,0.78))
axes1.add_artist(first_legend)
axes1.add_artist(two_legend)

axes1.set_title("Normal scale")


line1, = axes2.plot(x, x**2,  label ="Line 1")
line2, = axes2.plot(x, np.exp(x),  label ="Line 2")

# 
first_legend = axes2.legend(handles =[line1], loc = (0.1,0.88))
axes2.add_artist(first_legend)
axes2.legend(handles =[line2], loc = (0.1,0.78))

axes2.get_legend().set_visible(True)


axes2.set_yscale("log")
axes2.set_title("Logarithmic scale (y)");
```

```python
alpha = 0.7
phi_ext = 2 * np.pi * 0.5

def flux_qubit_potential(phi_m, phi_p):
    return 2 + alpha - 2 * np.cos(phi_p) * np.cos(phi_m) - alpha * np.cos(phi_ext - 2*phi_p)

phi_m = np.linspace(0, 2*np.pi, 100)
phi_p = np.linspace(0, 2*np.pi, 100)

#         
X,Y = np.meshgrid(phi_p, phi_m)
Z = flux_qubit_potential(X, Y)
fig, ax = plt.subplots()

p = ax.pcolor(X/(2*np.pi), Y/(2*np.pi), Z, cmap=plt.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max())

# ,          .    ,       ,       .
cb = fig.colorbar(p, ax=ax)
fig, ax = plt.subplots()

cn = ax.contour(Z, cmap=plt.cm.RdBu, vmin=abs(Z).min(), vmax=abs(Z).max(), extent=[0, 1, 0, 1])
```

```python
fig = plt.figure(figsize=(14,6))

# `ax`    3D,     projection='3d'   add_subplot
ax = fig.add_subplot(1, 2, 1, projection='3d')

p = ax.plot_surface(X, Y, Z, rstride=4, cstride=4, linewidth=0)

# surface_plot       colorbar
ax = fig.add_subplot(1, 2, 2, projection='3d')
#antialiased=False - 
#rstride cstride - 
p = ax.plot_surface(X, Y, Z, rstride=1, cstride=1, cmap=plt.cm.coolwarm, linewidth=1, antialiased=True)
#shrink -  
cb = fig.colorbar(p, shrink=0.5)
```
